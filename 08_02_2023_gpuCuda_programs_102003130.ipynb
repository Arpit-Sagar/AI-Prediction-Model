```python
%%cu
#include <stdio.h>
#include <stdlib.h>


#include <cuda_runtime.h>

#define THREADS_PER_BLOCK 1024

__global__ void mergeSort(int *arr, int *temp, int n) {
    int tid = blockIdx.x * blockDim.x + threadIdx.x;
    int left = tid * 2;
    int right = tid * 2 + 1;
    int mid = (left + right) / 2;

    if (left < n && right < n) {
        mergeSort<<<1,1>>>(arr, temp, left);
        mergeSort<<<1,1>>>(arr, temp, right);
        cudaDeviceSynchronize();
        int i = left, j = mid, k = 0;
        while (i < mid && j < right) {
            if (arr[i] < arr[j]) {
                temp[k++] = arr[i++];
            } else {
                temp[k++] = arr[j++];
            }
        }
        while (i < mid) {
            temp[k++] = arr[i++];
        }
        while (j < right) {
            temp[k++] = arr[j++];
        }
        for (i = 0; i < k; i++) {
            arr[left + i] = temp[i];
        }
    }
}

int main() {
    int n = 1024;
    int *arr = (int *)malloc(n * sizeof(int));
    int *temp = (int *)malloc(n * sizeof(int));
    int *d_arr, *d_temp;

    cudaMalloc(&d_arr, n * sizeof(int));
    cudaMalloc(&d_temp, n * sizeof(int));

    for (int i = 0; i < n; i++) {
        arr[i] = rand() % n;
    }

    cudaMemcpy(d_arr, arr, n * sizeof(int), cudaMemcpyHostToDevice);
    cudaMemcpy(d_temp, temp, n * sizeof(int), cudaMemcpyHostToDevice);

    mergeSort<<<n / THREADS_PER_BLOCK, THREADS_PER_BLOCK>>>(d_arr, d_temp, n);
    cudaDeviceSynchronize();

    cudaMemcpy(arr, d_arr, n * sizeof(int), cudaMemcpyDeviceToHost);

    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n");

    free(arr);
    free(temp);
    cudaFree(d_arr);
    cudaFree(d_temp);

    return 0;
}
```

```python
%%cu
//ADD two numbers using cuda
#include <iostream>
#include "cuda_runtime.h"
#include "device_launch_parameters.h"
#include <curand_kernel.h>
#include <ctime>
__global__ void add(int *a, int *b,int *c) //Kernel Definition
{
    *c = *a + *b;
    //printf(*a, *b, *c);
}
int main()
{
  int a, b, c; // host copies of a, b, c
  int *d_a, *d_b, *d_c; // device copies of a, b, c
  int size = sizeof(int);
  // Allocate space for device copies of a, b, c
  cudaMalloc((void **)&d_a, size);
  cudaMalloc((void **)&d_b, size);
  cudaMalloc((void **)&d_c, size);
  // Setup input values
  a = 2;
  b = 7;
  // Copy inputs to device
  cudaMemcpy(d_a, &a, size, cudaMemcpyHostToDevice);
  cudaMemcpy(d_b, &b, size, cudaMemcpyHostToDevice);
  // Launch add() kernel on GPU
   add<<<1,1>>>(d_a, d_b, d_c);
  // Copy result back to host
  cudaMemcpy(&c, d_c, size, cudaMemcpyDeviceToHost);
  printf("The answer is ");
  printf("%d",c);
  // Cleanup
  cudaFree(d_a); cudaFree(d_b); cudaFree(d_c);
  return 0;
}
```

```python
#kernel define
## Kernel - Adding two matrices MatA and MatB
%%cu
#include<stdio.h>
#include<cuda.h>
__global__ void MatAdd(float MatA[N][N], float MatB[N][N],
float MatC[N][N])
{
    int i = blockIdx.x * blockDim.x + threadIdx.x;
    int j = blockIdx.y * blockDim.y + threadIdx.y;
    if (i < N && j < N)
        MatC[i][j] = MatA[i][j] + MatB[i][j];
}

int main()
{

    // Matrix addition kernel launch from host code
    dim3 threadsPerBlock(16, 16);
    dim3 numBlocks((N + threadsPerBlock.x -1) / threadsPerBlock.x, (N+threadsPerBlock.y -1) / threadsPerBlock.y);
    MatAdd<<<numBlocks, threadsPerBlock>>>(MatA, MatB, MatC);

}
```

```python
%%cu
#include <stdio.h>
#include <stdlib.h>
#include<cuda.h>
#define N 2

__global__ void MatAdd(int A[][N], int B[][N], int C[][N]){
           int i = threadIdx.x;
           int j = threadIdx.y;

           C[i][j] = A[i][j] + B[i][j];
       }


int main(){

int A[N][N] = {{1,2},{3,4}};
int B[N][N] = {{5,6},{7,8}};
int C[N][N] = {{0,0},{0,0}};

int (*pA)[N], (*pB)[N], (*pC)[N];

cudaMalloc((void**)&pA, (N*N)*sizeof(int));//pA allocated on GPU
cudaMalloc((void**)&pB, (N*N)*sizeof(int));
cudaMalloc((void**)&pC, (N*N)*sizeof(int));

cudaMemcpy(pA, A, (N*N)*sizeof(int), cudaMemcpyHostToDevice);//CPU TO GPU COPIED
cudaMemcpy(pB, B, (N*N)*sizeof(int), cudaMemcpyHostToDevice);
cudaMemcpy(pC, C, (N*N)*sizeof(int), cudaMemcpyHostToDevice);

int numBlocks = 1;
dim3 threadsPerBlock(N,N);
MatAdd<<<numBlocks,threadsPerBlock>>>(pA,pB,pC);

cudaMemcpy(C, pC, (N*N)*sizeof(int), cudaMemcpyDeviceToHost);//GPU TO CPU COPIED

int i, j; printf("C = \n");
for(i=0;i<N;i++){
    for(j=0;j<N;j++){
        printf("%d ", C[i][j]);
    }
    printf("\n");
}

cudaFree(pA);
cudaFree(pB);
cudaFree(pC);
printf("\n");

return 0;
}
```

```python
%%cu
#include <stdio.h>
#ifdef _OPENACC
#include <openacc.h>
#endif

#define NX 102400

int main(void)
{
    double vecA[NX], vecB[NX], vecC[NX];
    double sum;
    int i;

    /* Initialization of the vectors */
    for (i = 0; i < NX; i++) {
        vecA[i] = 1.0 / ((double) (NX - i));
        vecB[i] = vecA[i] * vecA[i];
    }

    /*
     * Implement vector addition on device with OpenACC
     * vecC = vecA + vecB
     */
    for (i = 0; i < NX; i++) {
       vecC[i] = vecA[i] + vecB[i];
    }

    sum = 0.0;
    /* Compute the check value */
    for (i = 0; i < NX; i++) {
        sum += vecC[i];
    }
    printf("Reduction sum: %18.16f\n", sum);

    return 0;
}
```

```python
//matrix addition
%%cu
#include<iostream>
#include<cuda.h>
#include<cuda_runtime.h>

#define N 3

using namespace std;

__global__ void matrixAdd(int (*a)[N], int (*b)[N], int (*c)[N]){

	int i = blockIdx.x;
	int j = threadIdx.x;
	c[i][j] = a[i][j] + b[i][j];

	printf("\nblockIdx.x: %d\ttheradIdx.x: %d\tthreadIdx.y: %d",blockIdx.x,threadIdx.x,threadIdx.y);
}

int main(){

	int a[N][N] = {{1,2,3},{4,5,6},{7,8,9}};
	int b[N][N] = {{1,2,3},{4,5,6},{7,8,9}};

	int size = N * N * sizeof(int);

	int c[N][N];

	int (*dev_a)[N], (*dev_b)[N], (*dev_c)[N];

	cudaMalloc(&dev_a, size);
	cudaMalloc(&dev_b, size);
	cudaMalloc(&dev_c, size);

	cudaMemcpy(dev_a, a, size, cudaMemcpyHostToDevice);
	cudaMemcpy(dev_b, b, size, cudaMemcpyHostToDevice);
	cudaMemcpy(dev_c, c, size, cudaMemcpyHostToDevice);

	dim3 TPB(N,N);
	matrixAdd<<<1,TPB>>>(dev_a, dev_b, dev_c);

	matrixAdd<<<3,3>>>(dev_a, dev_b, dev_c);

	cudaMemcpy(a, dev_a, size, cudaMemcpyDeviceToHost);
	cudaMemcpy(b, dev_b, size, cudaMemcpyDeviceToHost);
	cudaMemcpy(c, dev_c, size, cudaMemcpyDeviceToHost);

	cudaDeviceSynchronize();

	cout<<endl;
	for(int i=0; i<N; i++){
		for(int j=0; j<N; j++){
			cout<<c[i][j]<<"\t";
		}
		cout<<endl;
	}


	cudaFree(dev_a);
	cudaFree(dev_b);
	cudaFree(dev_c);

	return 0;
}
```

MATRIX MULTIPLICATION

```python
%%cu
#include<openacc.h>
#include<stdio.h>

void matrix_mul(int n, float *A, float *B, float *C)
{
    #pragma acc parallel loop collapse(2)
    for (int i = 0; i < n; ++i) {
        for (int j = 0; j < n; ++j) {
            float sum = 0.0;
            for (int k = 0; k < n; ++k) {
                sum += A[i * n + k] * B[k * n + j];
            }
            C[i * n + j] = sum;
        }
    }
}

int main()
{
    int n = 1024; // Matrix size
    float *A = (float *)malloc(n * n * sizeof(float));
    float *B = (float *)malloc(n * n * sizeof(float));
    float *C = (float *)malloc(n * n * sizeof(float));

    // Initialize matrices A and B with random values
    for (int i = 0; i < n * n; ++i) {
        A[i] = (float)rand() / RAND_MAX;
        B[i] = (float)rand() / RAND_MAX;
    }

    // Compute matrix multiplication using OpenACC
    #pragma acc data copyin(A[0:n*n], B[0:n*n]) copyout(C[0:n*n])
    {
        matrix_mul(n, A, B, C);
    }

    // Print result
    printf("Matrix C:\n");
    for (int i = 0; i < n; ++i) {
        for (int j = 0; j < n; ++j) {
            printf("%.2f ", C[i * n + j]);
        }
        printf("\n");
    }

    free(A);
    free(B);
    free(C);

    return 0;
}
```
